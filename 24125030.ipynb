```python
import pandas as pd
from IPython.display import Markdown, display

file = "24125030.xlsx"
df = pd.read_excel(file, header=None)

display(Markdown(f"## Đây là bài tập của sinh viên: {df.iloc[0, 1]} - {df.iloc[1, 1]} - {df.iloc[2, 1]}"))
```

```python
import matplotlib.pyplot as plt
mssv = str(df.iloc[0, 1])
last_digit = int(mssv[-1])
data = pd.read_excel(file, skiprows=4, usecols="A:D")
avg_population = data.groupby("Quốc gia")["Dân số"].mean() / 1_000_000
plt.figure(figsize=(12, 7))
colors = plt.cm.viridis([i / len(avg_population) for i in range(len(avg_population))])

if last_digit % 2 == 0:
    avg_population.sort_values().plot(kind="barh", color=colors)
    plt.ylabel("Quốc gia")
    plt.xlabel("Dân số trung bình (2019–2023)")
    ax = plt.gca()
    ax.ticklabel_format(style='plain', axis='x')
    for i, v in enumerate(avg_population.sort_values()):
        ax.text(v + 0.4, i, f"{v:.1f}", va="center", fontsize=9)
    ax.text(1.005, -0.06, "Triệu người", transform=ax.transAxes, fontsize=10, fontweight="bold")
else:
    avg_population.plot(kind="bar", color=colors)
    plt.xlabel("Quốc gia")
    plt.ylabel("Dân số trung bình (2019–2023)")
    ax = plt.gca()
    ax.ticklabel_format(style='plain', axis='y')
    for i, v in enumerate(avg_population):
        ax.text(i, v + 0.4, f"{v:.1f}", ha="center", fontsize=9)
    ax.text(-0.06, 1.01, "Triệu người", transform=ax.transAxes, fontsize=10, fontweight="bold")

plt.title("Trung bình dân số các nước trong năm năm từ 2019–2023")
plt.tight_layout()
plt.show()
```
