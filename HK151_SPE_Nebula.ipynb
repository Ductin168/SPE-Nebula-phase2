## System Performance Steps

### 1. Define Goal and System
####1.1. State goal
Hệ thống được lựa chọn để mô phỏng lại hàng đợi của một cơ sở nha khoa, mục tiêu là đánh giá hiệu suất phục vụ, khám chữa bệnh của từng hàng đợi tương ứng với từng khu vực được chia trong cơ sở bằng mô hình hàng đợi M/M/c. Các mục tiêu chính bao gồm:
- Mô phỏng: Tạo một mô phỏng chi tiết về hệ thống hàng đợi của cơ sở nha khoa, với nhiều khu vực khác nhau.
- Phân tích: Phân tích kết quả dựa trên mô phỏng thu được như số lượng công việc hiện thực được trong quá trình mô phỏng, thông tin chi tiết, yếu tố khác nhau ảnh hưởng tới sự lưu thông của bệnh nhân trong cơ sở cũng như thời gian phải chờ đợi
- Đánh giá: Đánh giá toàn diện hệ thống được mô phỏng để qua đó tìm hiểu hiệu suất đạt được của hệ thống, hiệu quả sử dụng các server.
####1.2. Define system
#####1.2.1. Giới thiệu các thành phần hệ thống
Một cơ sở nha khoa chuyên điều trị khám răng thông thường và các tiểu phẫu nha (minor surgery) được mô phỏng theo mô hình hàng đợi M/M/c, gồm các khu vực chính sau:

Khu vực tiếp (Reception Queing):
-   Chức năng: Bệnh nhân đến đăng ký, khai báo triệu chứng, phân loại loại dịch vụ (khám thường, cạo vôi, nhổ răng tiểu phẫu, nội nha, v.v.), in số thứ tự và hướng dẫn tới khu vực tiếp theo.
-   Nhân sự: Lễ tân/điều dưỡng phụ trách phân loại, hẹn phòng, in vé, kiểm tra bảo hiểm (BHYT) sơ bộ.
-   Quy trình đầu: Đây là điểm vào hệ thống — thời điểm bắt đầu mô phỏng cho từng bệnh nhân.

Khu vực sàn lọc / tiểu phẫu (Pre-op / Testing Queing)
-   Chức năng: Những bệnh nhân có chỉ định cần xét nghiệm (ví dụ: huyết học, đông máu) hoặc cần sàng lọc tiền phẫu (hỏi tiền sử, đo huyết áp) trước khi thực hiện thủ thuật. Một số dịch vụ khám thông thường không cần qua bước này.
-   Nhân sự: Điều dưỡng, kỹ thuật viên xét nghiệm.
-   Lưu ý: Chỉ một phần bệnh nhân (theo xác suất) được chuyển đến khu này.

Khu vực khám / điều trị (Dental treatment / Minor surgery Queuing)
-   Chức năng: Khu chính gồm nhiều ghế nha (multiple dentist chair) / phòng nhỏ cho điều trị và tiểu phẩu. Bệnh nhân chờ tới lượt ghế/BS khám, thực hiện can thiệp nha hoặc tiểu phẫu (ví dụ nhổ răng phức tạp, khâu mềm ). Sau khi khám/tiểu phẫu, bác sĩ có thể kê đơn thuốc hoặc chỉ định tái khám.
-   Nhân sự: Bác sĩ nha khoa / thầy thuốc (các server trong mô hình M/M/c), trợ thủ nha khoa, y tá hỗ trợ phẫu thuật.

Khu vực thanh toán và nhận thuốc (Payment / Pharmacy Queuing)
-   Chức năng: Thanh toán dịch vụ, kiểm tra BHYT chính thức, phát thuốc theo đơn. Sau khi hoàn tất thanh toán và nhận thuốc, bệnh nhân rời hệ thống — kết thúc mô phỏng.
##### 1.2.2. Các giả sử của hệ thống

Để thực hiện mô phỏng và đánh giá hiệu năng, ta đặt các giả sử và ràng buộc sau cho mỗi hàng đợi:

Quy trình đến (Arrival Process)
-   Bệnh nhân đến trung tâm theo một quá trình Poisson với tốc độ trung bình λ (số bệnh nhân/đơn vị thời gian). Thời gian giữa các lần đến là biến ngẫu nhiên độc lập, phân phối mũ (Exponential).
-   Tại thời điểm bệnh nhân bước vào Reception được coi là bắt đầu quan sát/mô phỏng cho bệnh nhân đó.
Thời gian phục vụ (Service time)
-   Thời gian phục vụ tại mỗi server (lễ tân, kỹ thuật viên xét nghiệm, bác sĩ/ghế nha, thu ngân, dược sĩ) được giả sử là các biến ngẫu nhiên độc lập, phân phối mũ với trung bình 1/μi khác nhau tùy khu vực i.
-   Thời gian khử khuẩn/chuẩn bị ghế cũng được coi là thời gian phục vụ (có thể mô hình hóa như là một server riêng hoặc gộp vào thời gian rời giường).
Mô hình hàng đợi (System Capacity & Discipline)
-   Mỗi khu vực tuân theo mô hình M/M/c (c servers) với sức chứa hệ thống là vô hạn (infinite capacity waiting room).
-   Nguyên tắc phục vụ là FCFS (First Come, First Served) trong mỗi queue. (Có thể mở rộng sau này bằng ưu tiên cho trường hợp cấp cứu — nếu muốn, mô phỏng thêm lớp ưu tiên).
Quy tắc chuyển tuyến (Routing & Probabilities)
-   Sau Reception, bệnh nhân được chuyển tới khu Testing với xác suất *P*test (tùy loại dịch vụ). Ngược lại, với xác suất 1 − *P*test bệnh nhân đi thẳng tới khu khám.
-   Sau khám: với xác suất *P*surgery bệnh nhân cần tiểu phẫu (thời gian phục vụ lớn hơn trung bình), với xác suất khác là khám/điều trị thông thường. Một số ca sau phẫu thuật vào Recovery với xác suất *P*recovery.
-   Sau khi hoàn tất (khám/tiểu phẫu và recovery nếu có), bệnh nhân sẽ được chuyển tới Payment.

Sức chứa hệ thống
-   Hệ thống giả sử không giới hạn số lượng bệnh nhân chờ (infinite buffer), vì mục tiêu là đo tải và thời gian chờ.
Quy tắc khác và phạm vi mô phỏng
-   Mô phỏng chỉ bắt đầu khi bệnh nhân vào Reception; các hoạt động ngoài (hẹn, gọi điện nhắc lịch, v.v.) nằm ngoài phạm vi.
-   Mô phỏng kết thúc khi bệnh nhân đã hoàn tất thanh toán và nhận thuốc.
-   Các biến cố hiếm như tai biến nặng, chuyển viện cấp cứu, hoặc chậm trễ do thiếu thiết bị chuyên sâu được coi là ngoài phạm vi.
-   Không mô phỏng hành vi cá nhân phức tạp như rời hàng (balking) hoặc hủy hẹn giữa chừng (reneging) ban đầu.

#### 1.3. Đặc tả thông số mô hình
* Đối với đề tài hệ thống xếp hàng của một nha khoa, nhóm triển khai bằng các Kendall's notation. Mô hình được nhóm sử dụng sẽ là M/M/c.
* Thời gian bệnh nhân đến (Arrival process): Đối với mô phỏng đây sẽ là thời gian bệnh nhân bắt đầu đến quầy tiếp tân để lấy số báo danh cho việc khám. Thời gian giữa các hàng đợi là độc lập, ngẫu nhiên và tuân theo phân phối mũ với giá trị M(Markovian).
* Thời gian phục vụ (service times): Đây là thời gian khám, chữa bệnh cho bệnh nhân tuân theo nguyên tắc hệ số mũ với giá trị M. Do đó, thời gian khám chữa bệnh cũng độc lập, ngẫu nhiên.
* Số lượng server: Là số lượng các thành phần hay đơn vị của các phòng ban nhầm thực hiện khám, tiểu phẩu, thanh toán viện phí cho bệnh nhân. Đối với mô hình này là c với c > 1.
* Sức chứa hệ thống (System Capacity): Đây là tổng số bệnh nhân tối đa hệ thống có thể tiếp nhận. Đối với bài mô hình này là vô hạn.
* Kích thước nguồn (population size): Đây là kích thước nguồn khách. Đối với mô phỏng này là vô hạn.
* Quy tắc hàng đợi: Đây là quy ước cách hệ thống giải quyết vấn đề các hàng đợi hay bệnh nhân. Đối với hệ thống phòng khám nha khoa sẽ là FCFS( First Come, First Served).
* Hệ thống mô tả quá trình 1 bệnh nhân thực hiện lấy số thứ tự từ quầy tiếp tân đến khi hoàn thành việc thanh toán viện phí. Trong đó, bệnh nhân có thể được luân chuyển giữa các phòng khám, tiểu phẫu hay lấy thuốc ngẫu nhiên.
* Bênh cạnh đó mô hình có thêm thực hiện đặt lịch thực hiện tiểu phẩu sau khi khám.
### 2. Service and Outcomes
#### 2.1. Services
Dựa trên các thành phần và giả sử đã đề cập, các dịch vụ do hệ thống cung cấp bao gồm:
- Khu vực tiếp nhận, bệnh nhân được lấy số thứ tự để chờ. Tại đây bệnh nhân được nhận số thứ tự để chờ đợi, nhân viên lễ tân (điều dưỡng) sẽ tư vấn về nhu cầu kiểm tra, khám chữa và điều trị của bệnh nhân, hướng dẫn họ đến khu vực phù hợp.
- Khi vào khu vực khám bệnh, bệnh nhân được dẫn đến các phòng khám tương ứng với vấn đề của mình hoặc ưu tiên đến khu vực xét nghiệm nếu có yêu cầu. Việc lựa chọn loại hình khám là ngẫu nhiên với tỷ lệ phân bố sau:
  + Bệnh nhân chọn khám bệnh A (không cần xét nghiệm), lúc này họ sẽ tiến đến hàng đợi khám chữa bệnh (Medical Examination Queuing) với tỷ lệ xảy ra là a.
  + Bệnh nhân chọn khám bệnh B (yêu cầu xét nghiệm trước), họ sẽ được chuyển đến hàng đợi xét nghiệm (Testing Queuing). Sau khi hoàn tất xét nghiệm, bệnh nhân được tiếp tục chuyển sang hàng đợi khám chữa bệnh (Medical Examination Queuing) với tỷ lệ xảy ra là b.
- Tại hàng đợi khám chữa bệnh (Medical Examination Queuing), bệnh nhân được phân bổ đến bác sĩ phù hợp dựa trên chỉ định ban đầu. Quy trình tiếp theo diễn ra theo 2 hướng:
  + Bệnh nhân có thể trở lại để kiểm tra một vấn đề khác với tỉ lệ c.
  + Sau khi hoàn thành, bệnh nhân không cần khám thêm và di chuyển trực tiếp đến hàng đợi thanh toán cùng nhận thuốc (Paying Queuing) với tỉ lệ d.
- Khi đã kết thúc quy trình khám, bệnh nhân đến khu vực thanh toán và nhận thuốc (Payment Queuing). Sau khi lấy được đơn thuốc theo chỉ định của bác sĩ, bệnh nhân được coi là hoàn tất và rời khỏi hệ thống.
#### 2.2 Outcomes
Từ các dịch vụ khám chữa bệnh do hệ thống này cung cấp, có nhiều kết quả có thể xảy ra như sau:
- Ở khu vực tiếp đón: Bệnh nhân nhận được phiếu hẹn thành công và di chuyển vào khu vực khám, dẫn đến hai khả năng chính:
  + Chuyển sang khu vực xét nghiệm để thực hiện thủ tục cần thiết.
  + Tiến thẳng đến khu vực khám chữa bệnh mà không cần bước trung gian.

  Các điểm nghẽn tiềm ẩn: Hệ thống bị quá tải do số lượng bệnh nhân dồn dập, dẫn đến không thể tiếp nhận thêm; hoặc bệnh nhân không hài lòng với lời tư vấn và quyết định rời đi.
- Ở khu vực khám chữa bệnh: Sau khi hoàn tất buổi khám, bệnh nhân sẽ đối mặt với hai lựa chọn:
  + Di chuyển trực tiếp đến khu vực thanh toán và nhận thuốc.
  + Mong muốn kiểm tra thêm vấn đề sức khỏe khác, nên quay lại hàng đợi để chờ lượt tiếp theo.
- Ngoài ra, bệnh nhân có nguy cơ phải chờ đợi kéo dài trước khi được phục vụ, gây ảnh hưởng đến trải nghiệm tổng thể.

### 4. List System Parameters
Hệ thống sẽ dùng các thông số để thay đổi thay đổi hành vi cũng nhưng số lượng bệnh nhân được tạo ra, thời gian và hiệu suất phục vụ của các đơn vị:

  **Workload parameter**

* Số lượng server(c): Như đã nhắc đến trong mục [1.3](#scrollTo=vZhnVr2dBMuW), thông số ảnh hưởng tới tốc độ phục vụ của hệ thống và khả năng xử lý đồng thời. Đối với mô hình của bệnh viện nha khoa sẽ là:
    - Số lượng lễ tân tại khu vực tiếp nhận.
    - Số lượng bác sĩ thực hiện khám bệnh cũng như xét nghiệm, chụp X quang tại phòng khám. Mỗi bác sĩ sẽ tương ứng với một phòng khám trong mô hình sẽ tương ứng với 1 phòng khám bệnh.
    - Số lượng phòng tiểu phẫu dùng để thực hiện tiễu phẫu cho các bệnh nhân.
    - Số lượng điều dưỡng tại khu vực thanh toán và lấy thuốc.
* Arrival Rate ($ \lambda $): Tỷ lệ bệnh nhân đến hệ thống khám chữa bệnh, là số lượng bệnh nhân trung bình trên một đơn vị thời gian.
* Service Rate (μ): Tỷ lệ phục vụ bệnh nhân của hệ thống, đây là giá trị ảnh hưởng đến thời gian thực hiện tiểu phẩu, khám bệnh của một bệnh nhân.

  Các parameter trên đều có thể thay đổi để tạo dựng các tình huống (workload) khác nhau cho việc mô phỏng. Các tình huống này, đối với mô hình nha khoa sẽ là thời điểm quá tải, trống hay là tiền đề đặt ra quy mô của bệnh viện.

* Thời gian khách hàng phải chờ đợi: Đây là thông số để đánh giá tượng quan giữa hiệu năng làm việc của hệ thông bệnh viện so với tỷ lệ bệnh nhân chờ khám. Thông số sẽ được đánh giá ở từng khu vực và được tính tổng thể nhằm cho biết các đơn vị cần cải thiện và hiệu suất cải thiện.

**System parameter**

* Thời gian phục vụ: Đây sẽ là thời gian để một bệnh nhân dùng dịch vụ khám chữa bệnh tại từng hàng đợi và tổng thể trong một lần đến khám chữa bệnh tại bệnh viện. (Không tính thời gian kể từ lúc bệnh nhân đặt lịch thực hiện tiểu phẫu).


Parameters


```python
# Simulation parameters

# Global RNG seed (set to None for non-deterministic runs)
RANDOM_SEED = 123

# Arrival process
ARRIVAL_RATE = 0.3       # lambda (arrivals per time unit) - exponential interarrival

# Service rates (mu) for each node (exponential service times)
MU_RECEPTION = 0.5
MU_TESTING   = 0.4
MU_EXAM      = 0.3
MU_PAYMENT   = 0.6

# Number of servers (c) per node
SERVERS_RECEPTION = 2
SERVERS_TESTING   = 1
SERVERS_EXAM      = 3
SERVERS_PAYMENT   = 1

# Routing probabilities
P_TEST    = 0.3   # probability, after reception, patient goes to testing (otherwise straight to exam)
P_REEXAM  = 0.1   # probability, after an exam, the patient goes back to exam again (feedback)

# Simulation control
SIM_TIME = 20000.0   # total simulated time
WARMUP_TIME = 0.0   # optional warmup period if need to avoid transient period

# Toggle for printing real-time events
ENABLE_TRACE = True

def trace(msg: str):
    """Print if tracing is enabled."""
    if ENABLE_TRACE:
        print(msg)
```

Hook interfaces for metrics and analytics

```python
# Simple hook/callback structure so metrics/analytics can attach functions.
# Each hook is optional (set to a callable that accepts specific args), or None.

# Hook signatures (expected):
#   on_arrival(patient_id:int, now:float, node_name:str)
#   on_service_start(patient_id:int, now:float, node_name:str)
#   on_service_end(patient_id:int, now:float, node_name:str)
#   on_departure(patient_id:int, now:float)

# Default no-op callbacks:
def _noop(*_, **__):
    return None

class Callbacks:
    def __init__(self,
                 on_arrival: Optional[Callable[[int, float, str], Any]] = None,
                 on_arrival_node: Optional[Callable[[int, float, str], Any]] = None,
                 on_service_start: Optional[Callable[[int, float, str], Any]] = None,
                 on_service_end: Optional[Callable[[int, float, str], Any]] = None,
                 on_departure: Optional[Callable[[int, float], Any]] = None):

        self.on_arrival = on_arrival or _noop
        self.on_arrival_node = on_arrival_node or _noop
        self.on_service_start = on_service_start or _noop
        self.on_service_end = on_service_end or _noop
        self.on_departure = on_departure or _noop
```

Queue node

```python
class QueueNode:
    """
    Generic M/M/c node (simulation core) with optional tracing.
    """
    def __init__(self, env, name, n_servers, service_rate, callbacks=None):
        self.env = env
        self.name = name
        self.n_servers = int(n_servers)
        self.service_rate = float(service_rate)
        self.resource = simpy.Resource(env, capacity=self.n_servers)
        self.callbacks = callbacks or Callbacks()

    def _draw_service_time(self):
        return random.expovariate(self.service_rate)

    def serve(self, patient_id):
        arrival_to_node = self.env.now

        trace(f"[t={self.env.now:.2f}] Patient {patient_id} ARRIVED at {self.name} queue.")

        try:
            self.callbacks.on_arrival_node(patient_id, arrival_to_node, self.name)
        except:
            pass

        with self.resource.request() as req:
            yield req
            start = self.env.now

            trace(f"[t={start:.2f}] Patient {patient_id} STARTS service at {self.name} (servers busy={self.resource.count}).")

            try:
                self.callbacks.on_service_start(patient_id, start, self.name)
            except:
                pass

            service_duration = self._draw_service_time()
            yield self.env.timeout(service_duration)

            end = self.env.now
            trace(f"[t={end:.2f}] Patient {patient_id} FINISHES service at {self.name}. Duration={service_duration:.2f}")

            try:
                self.callbacks.on_service_end(patient_id, end, self.name)
            except:
                pass
```

Build simulation system

```python
def build_system(env: simpy.Environment, callbacks: Optional[Callbacks] = None,
                 mu_reception=MU_RECEPTION, mu_testing=MU_TESTING,
                 mu_exam=MU_EXAM, mu_payment=MU_PAYMENT,
                 c_reception=SERVERS_RECEPTION, c_testing=SERVERS_TESTING,
                 c_exam=SERVERS_EXAM, c_payment=SERVERS_PAYMENT) -> Dict[str, QueueNode]:

    callbacks = callbacks or Callbacks()

    nodes = {
        "reception": QueueNode(env, "reception", c_reception, mu_reception, callbacks),
        "testing":   QueueNode(env, "testing",   c_testing,   mu_testing,   callbacks),
        "exam":      QueueNode(env, "exam",      c_exam,      mu_exam,      callbacks),
        "payment":   QueueNode(env, "payment",   c_payment,   mu_payment,   callbacks),
    }
    return nodes
```

Main simulation logic (patients process)

```python
def patient_process(env, patient_id, nodes, callbacks=None):
    callbacks = callbacks or Callbacks()
    trace(f"\n=== Patient {patient_id} ENTERS system at t={env.now:.2f} ===")
    try:
        callbacks.on_arrival(patient_id, env.now, "system_entry")
    except:
        pass

    # 1. Reception
    trace(f"Patient {patient_id} going to RECEPTION.")
    yield env.process(nodes["reception"].serve(patient_id))

    # 2. Routing: testing or straight to exam
    if random.random() < P_TEST:
        trace(f"Patient {patient_id} routed to TESTING.")
        yield env.process(nodes["testing"].serve(patient_id))
    else:
        trace(f"Patient {patient_id} SKIPS testing and goes to EXAM.")

    # 3. Examination, possibly multiple times (feedback loop)
    while True:
        trace(f"Patient {patient_id} entering EXAMINATION.")
        yield env.process(nodes["exam"].serve(patient_id))

        if random.random() < P_REEXAM:
            trace(f"Patient {patient_id} requires RE-EXAM. Returning to EXAM queue.")
        else:
            break

    # 4. Payment
    trace(f"Patient {patient_id} going to PAYMENT.")
    yield env.process(nodes["payment"].serve(patient_id))

    trace(f"=== Patient {patient_id} LEAVES system at t={env.now:.2f} ===\n")

    try:
        callbacks.on_departure(patient_id, env.now)
    except:
        pass
```

Arrival generator

```python
def arrival_generator(env: simpy.Environment, nodes: Dict[str, QueueNode],
                      callbacks: Optional[Callbacks] = None,
                      max_arrivals: Optional[int] = None,
                      arrival_rate: float = ARRIVAL_RATE):
    callbacks = callbacks or Callbacks()
    patient_id = 0
    while True:
        inter = random.expovariate(arrival_rate)  # FIXED: dùng tham số
        yield env.timeout(inter)

        env.process(patient_process(env, patient_id, nodes, callbacks))
        patient_id += 1

        if max_arrivals is not None and patient_id >= max_arrivals:
            break
```

Run simulation

```python
def run_simulation(sim_time: float = SIM_TIME, callbacks: Optional[Callbacks] = None,
                   warmup: float = WARMUP_TIME, max_arrivals: Optional[int] = None,
                   seed: Optional[int] = RANDOM_SEED,
                   arrival_rate=ARRIVAL_RATE,
                   mu_reception=MU_RECEPTION, mu_testing=MU_TESTING,
                   mu_exam=MU_EXAM, mu_payment=MU_PAYMENT,
                   c_reception=SERVERS_RECEPTION, c_testing=SERVERS_TESTING,
                   c_exam=SERVERS_EXAM, c_payment=SERVERS_PAYMENT):

    if seed is not None:
        random.seed(seed)

    env = simpy.Environment()

    nodes = build_system(
        env, callbacks,
        mu_reception=mu_reception, mu_testing=mu_testing,
        mu_exam=mu_exam, mu_payment=mu_payment,
        c_reception=c_reception, c_testing=c_testing,
        c_exam=c_exam, c_payment=c_payment
    )

    env.process(arrival_generator(env, nodes, callbacks=callbacks,
                                  max_arrivals=max_arrivals,
                                  arrival_rate=arrival_rate))

    env.run(until=sim_time)
    return env, nodes


# Example run (no metrics hooked)
if __name__ == "__main__":
    print("Running default workload test (no metrics)...\n")

    env, nodes = run_simulation(
        sim_time=200.0,
        seed=RANDOM_SEED,
        arrival_rate=ARRIVAL_RATE,
        mu_reception=MU_RECEPTION,
        mu_testing=MU_TESTING,
        mu_exam=MU_EXAM,
        mu_payment=MU_PAYMENT,
        c_reception=SERVERS_RECEPTION,
        c_testing=SERVERS_TESTING,
        c_exam=SERVERS_EXAM,
        c_payment=SERVERS_PAYMENT
    )

    print("\nSimulation finished (no metrics attached). Node servers and rates:")
    for name, node in nodes.items():
        print(f" - {name}: servers={node.n_servers}, mu={node.service_rate}")
```

Thu thập metrics

```python
import numpy as np

class MetricsRecorder:
    def __init__(self):
        self.entry_time = {}
        self.exit_time = {}

        self.arrival_times = {}
        self.service_start_times = {}

        self.queues = ["reception", "testing", "exam", "payment"]

        self.waiting_times  = { q: [] for q in self.queues }
        self.service_times  = { q: [] for q in self.queues }
        self.response_times = { q: [] for q in self.queues }

        self.busy_time = { q: 0.0 for q in self.queues }

        self.queue_length = { q: 0 for q in self.queues }
        self.max_queue_length = { q: 0 for q in self.queues }

        self.queue_length_area = { q: 0.0 for q in self.queues }
        self.last_event_time = 0.0



    def _update_queue_length_area(self, now):
        dt = now - self.last_event_time
        if dt > 0:
            for q in self.queues:
                self.queue_length_area[q] += max(self.queue_length[q], 0) * dt
        self.last_event_time = now



    def on_arrival(self, pid, now, node_name):
        if node_name == "system_entry":
            self.entry_time[pid] = now


    def on_arrival_node(self, pid, now, node_name):
        self._update_queue_length_area(now)

        self.arrival_times[(pid, node_name)] = now
        self.queue_length[node_name] += 1

        if self.queue_length[node_name] > self.max_queue_length[node_name]:
            self.max_queue_length[node_name] = self.queue_length[node_name]


    def on_service_start(self, pid, now, node_name):
        self._update_queue_length_area(now)

        key = (pid, node_name)
        self.service_start_times[key] = now

        arr = self.arrival_times.get(key)
        if arr is not None:
            self.waiting_times[node_name].append(now - arr)

        # Prevent negative queue length
        self.queue_length[node_name] = max(0, self.queue_length[node_name] - 1)


    def on_service_end(self, pid, now, node_name):
        self._update_queue_length_area(now)

        key = (pid, node_name)
        start = self.service_start_times.get(key)

        if start is not None:
            service = now - start
            self.service_times[node_name].append(service)
            self.busy_time[node_name] += service

        arr = self.arrival_times.get(key)
        if arr is not None:
            self.response_times[node_name].append(now - arr)


    def on_departure(self, pid, now):
        self.exit_time[pid] = now


    def summary(self, sim_time, servers):

        completed = len(self.exit_time)

        system_times = [
            self.exit_time[pid] - t_in
            for pid, t_in in self.entry_time.items()
            if pid in self.exit_time
        ]

        avg_system_time = sum(system_times)/len(system_times) if system_times else 0

        utilization = {
            q: self.busy_time[q] / (sim_time * servers[q])
            for q in self.queues
        }

        throughput = completed / sim_time

        avg_Lq = {
            q: self.queue_length_area[q] / sim_time
            for q in self.queues
        }

        waiting_p90 = { q: (float(np.percentile(self.waiting_times[q], 90)) if self.waiting_times[q] else 0)
                        for q in self.queues }
        waiting_p95 = { q: (float(np.percentile(self.waiting_times[q], 95)) if self.waiting_times[q] else 0)
                        for q in self.queues }

        response_p90 = { q: (float(np.percentile(self.response_times[q], 90)) if self.response_times[q] else 0)
                        for q in self.queues }
        response_p95 = { q: (float(np.percentile(self.response_times[q], 95)) if self.response_times[q] else 0)
                        for q in self.queues }

        return {
            "waiting_times": self.waiting_times,
            "service_times": self.service_times,
            "response_times": self.response_times,

            "busy_time": self.busy_time,
            "utilization": utilization,

            "avg_queue_length": avg_Lq,
            "max_queue_length": self.max_queue_length,

            "patients_completed": completed,
            "avg_system_time": avg_system_time,
            "throughput": throughput,

            "waiting_p90": waiting_p90,
            "waiting_p95": waiting_p95,
            "response_p90": response_p90,
            "response_p95": response_p95
        }
```

Run workload

```python
def run_one_workload(
    lam,
    sim_time=20000,

    # service rates (mu)
    mu_reception=MU_RECEPTION,
    mu_testing=MU_TESTING,
    mu_exam=MU_EXAM,
    mu_payment=MU_PAYMENT,

    # number of servers (c)
    c_reception=SERVERS_RECEPTION,
    c_testing=SERVERS_TESTING,
    c_exam=SERVERS_EXAM,
    c_payment=SERVERS_PAYMENT
):
    global ENABLE_TRACE
    ENABLE_TRACE = False

    metrics = MetricsRecorder()

    cb = Callbacks(
        on_arrival=metrics.on_arrival,
        on_arrival_node=metrics.on_arrival_node,
        on_service_start=metrics.on_service_start,
        on_service_end=metrics.on_service_end,
        on_departure=metrics.on_departure
    )

    servers = {
        "reception": c_reception,
        "testing":   c_testing,
        "exam":      c_exam,
        "payment":   c_payment
    }

    run_simulation(
        sim_time=sim_time,
        callbacks=cb,
        arrival_rate=lam,

        mu_reception=mu_reception,
        mu_testing=mu_testing,
        mu_exam=mu_exam,
        mu_payment=mu_payment,

        c_reception=c_reception,
        c_testing=c_testing,
        c_exam=c_exam,
        c_payment=c_payment
    )

    summary = metrics.summary(sim_time, servers)


    print(f"\n================ WORKLOAD λ = {lam} ================\n")

    left  = []
    right = []

    left.append(f"Simulation time T         : {sim_time}")
    left.append(f"Patients completed (C)    : {summary['patients_completed']}")
    left.append(f"Throughput X = C/T        : {summary['throughput']:.4f}")
    left.append(f"Avg system time (Wsys)    : {summary['avg_system_time']:.4f}")
    left.append("")
    left.append("--- Average Waiting Time (Wq) ---")
    for q in metrics.queues:
        w = summary["waiting_times"][q]
        left.append(f"{q:<10}: {(sum(w)/len(w) if w else 0):.4f}")

    left.append("")
    left.append("--- P90 / P95 Waiting Time ---")
    for q in metrics.queues:
        left.append(f"{q:<10}: P90={summary['waiting_p90'][q]:.4f}   P95={summary['waiting_p95'][q]:.4f}")

    left.append("")
    left.append("--- Average Service Time (S) ---")
    for q in metrics.queues:
        s = summary["service_times"][q]
        left.append(f"{q:<10}: {(sum(s)/len(s) if s else 0):.4f}")

    right.append("--- Average Response Time (R) ---")
    for q in metrics.queues:
        r = summary["response_times"][q]
        right.append(f"{q:<10}: {(sum(r)/len(r) if r else 0):.4f}")

    right.append("")
    right.append("--- P90 / P95 Response Time ---")
    for q in metrics.queues:
        right.append(f"{q:<10}: P90={summary['response_p90'][q]:.4f}   P95={summary['response_p95'][q]:.4f}")

    right.append("")
    right.append("--- Utilization ρ ---")
    right.append("Formula: ρ = busy_time / (T × c)")
    for q in metrics.queues:
        busy = summary["busy_time"][q]
        c = servers[q]
        rho = summary["utilization"][q]
        right.append(f"{q:<10}: ρ={rho:.4f}   (busy={busy:.2f}, T={sim_time}, c={c})")

    right.append("")
    right.append("--- Empirical Queue Length Lq ---")
    for q in metrics.queues:
        right.append(f"{q:<10}: {summary['avg_queue_length'][q]:.4f}")

    right.append("")
    right.append("--- Max Queue Length Observed ---")
    for q in metrics.queues:
        right.append(f"{q:<10}: {summary['max_queue_length'][q]}")

    WIDTH = 55

    print("")
    for i in range(max(len(left), len(right))):
        L = left[i]  if i < len(left)  else ""
        R = right[i] if i < len(right) else ""
        print(f"{L:<{WIDTH}} | {R}")

    print("\n============================================================\n")

```

## I. Điểm mạnh nổi bật (Superior Points)
### 1. Mô hình hóa chính xác bài toán thực tế
Hệ thống đã tái hiện thành công quy trình khám chữa bệnh phức tạp tại nha khoa thông qua mạng lưới hàng đợi M/M/c đa tầng (Reception – Testing – Examination – Payment).
- Luồng di chuyển linh hoạt: Không chỉ là tuyến tính đơn giản, hệ thống xử lý tốt các kịch bản rẽ nhánh (Routing xác suất đi xét nghiệm) và vòng lặp phản hồi (Feedback loop - quay lại tái khám), phản ánh sát với hành vi thực tế của bệnh nhân.
- Cấu trúc đa server: Các node được thiết kế với số lượng tài nguyên khác nhau (đặc biệt là Exam và Reception), phù hợp với bài toán "multiple dentist chair/minor surgery system".

### 2. Năng lực phát hiện và giải quyết điểm nghẽn dựa trên dữ liệu
Mô hình chứng minh được giá trị thông qua việc nhận diện chính xác điểm nghẽn (Bottleneck) khi thay đổi tải trọng:
- Nhận diện Bottleneck: Tại mức tải λ = 0.4, hệ thống ổn định (thời gian chờ ~13s). Tuy nhiên, khi tăng lên λ = 0.6, mô phỏng chỉ ra sự "gãy đổ" tại nút Payment (Utilization ≈ 99%, Max Queue lên tới 156 người).
- So sánh giải pháp: Hệ thống cho phép đánh giá định lượng hiệu quả giữa hai chiến lược: Mở rộng quy mô (Tăng server) và Tăng năng suất (Tăng service rate). Kết quả cho thấy cả hai đều giải quyết được tắc nghẽn, đưa hệ thống về trạng thái ổn định, từ đó cung cấp cơ sở ra quyết định chính xác.

### 3. Hệ thống chỉ số đo lường toàn diện và chuyên sâu
Không chỉ dừng lại ở các chỉ số cơ bản (Waiting time, Utilization), hệ thống còn cung cấp các metric nâng cao hỗ trợ đánh giá Chất lượng dịch vụ (QoS):
- Percentile Metrics (P90/P95): Việc tích hợp các chỉ số phân vị giúp đánh giá trải nghiệm của nhóm khách hàng rủi ro nhất (5-10% bệnh nhân phải chờ lâu), thay vì chỉ nhìn vào con số trung bình có thể gây hiểu lầm.
- Đầy đủ các chiều đo: Bao gồm Throughput, Queue Length (Avg/Max), Response Time cho từng bộ phận riêng biệt.

## II. Các hạn chế (Limitations)
### 1. Giả định mô hình còn đơn giản so với biến thiên thực tế
- Static Workload: Tham số λ được giữ cố định trong suốt quá trình chạy, chưa phản ánh được tính chất "Dynamic Workload" của thực tế (VD: giờ cao điểm sáng/chiều hay sự thay đổi theo thời gian thực).
- Infinite Capacity: Hệ thống giả định hàng đợi vô hạn và bệnh nhân có sự kiên nhẫn tuyệt đối. Thực tế cần tính đến giới hạn không gian vật lý và hành vi rời bỏ hàng đợi khi chờ quá lâu.

### 2. Phạm vi thực nghiệm và độ sâu thống kê chưa cao
- Single Run Simulation: Mỗi kịch bản (Workload) chỉ được chạy một lần duy nhất. Việc thiếu các lần chạy lặp lại (Replications) khiến kết quả chưa tính toán được độ lệch chuẩn hay khoảng tin cậy (Confidence Interval), làm giảm tính thuyết phục về mặt thống kê.
- Chưa tối ưu toàn cục: Các kịch bản hiện tại chỉ mới giải quyết điểm nghẽn tại Payment. Sau khi Payment được giải quyết, nút Exam trở thành điểm bận nhất (ρ ≈ 0.74), nhưng chưa có kịch bản thử nghiệm tiếp theo để tối ưu hóa toàn diện mạng lưới.

### 3. Hạn chế về mặt Kỹ thuật và Trực quan hóa
- Thiếu trực quan hóa (Visualization): Kết quả đầu ra hoàn toàn là dạng văn bản (text/log), thiếu các biểu đồ (Chart/Graph) về biến thiên độ dài hàng đợi theo thời gian, gây khó khăn trong việc quan sát xu hướng tức thời.
- Vấn đề về Trace Log: Khi bật chế độ theo dõi chi tiết, lượng log in ra quá lớn gây nhiễu thông tin (Noise) và ảnh hưởng đến hiệu suất mô phỏng.
- Thiếu kiểm thử biên: Chưa có các test case cho các trường hợp tham số cực đoan (như λ quá lớn gây tràn bộ nhớ hoặc server = 0) để kiểm chứng độ bền vững của code.
